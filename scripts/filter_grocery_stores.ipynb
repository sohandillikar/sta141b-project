```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('../data/grocery_stores_v1.csv')
df.head()
```

```python
all_types = []

for types_str in df['types'].dropna():
    types = ast.literal_eval(types_str)
    all_types.extend(types)

all_types = list(set(all_types))

len(all_types)
```

```python
all_types
```

```python
def is_grocery_market(types_str):
    types = ast.literal_eval(types_str)
    return 'grocery_or_supermarket' in types or 'supermarket' in types

df_filtered = df[df['types'].apply(is_grocery_market)]

print(f'Old length: {len(df)}\nNew length: {len(df_filtered)}')
```

```python
df_filtered
```

```python
df_filtered.reset_index(drop=True, inplace=True)
df_filtered.to_csv('../data/grocery_stores_v3.csv')
```
